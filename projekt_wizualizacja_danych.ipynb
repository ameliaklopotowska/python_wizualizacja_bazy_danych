Ilość produktów każdej kategorii w sklepie

```python
#Łączę tabele kategorie z tabelą produkty, aby mieć tabele z nazwami poszczególnych kategorii
```

```python
df1 = produkty_kategorie.groupby(by='OPIS').count()['IDPRODUKTU'].reset_index().sort_values(by='IDPRODUKTU', ascending=False)
```

```python
#wykresy słupkowe matplotlib
plt.figure(figsize=(15,7))
plt.grid()
plt.bar(df1['OPIS'], df1['IDPRODUKTU'], color='blue')
plt.title('Ilość produktów w sklepie w poszczególnych kategoriach')
plt.xlabel('')
plt.ylabel('Ilość produktów')
plt.xticks(rotation=90)
plt.show()
```

Stan magazynowy produktów

```python
print(produkty['STANMAGAZYNU'].min(), produkty['STANMAGAZYNU'].mean(),produkty['STANMAGAZYNU'].max())
```

```python
produkty_kategorie = produkty_kategorie.sort_values(by='IDPRODUKTU')
```

```python
produkty_kategorie
```

```python
produkty_kategorie['STANMAGAZYNU'].values
```

```python
colors=[]
for i,r in produkty_kategorie.iterrows():
    if r['STANMAGAZYNU'] <=r['STANMINIMUM']:
        colors.append('r')
    else:
        colors.append('b')
# clrs = ['red' if (x <= 20) else 'blue' for x in produkty_kategorie['STANMAGAZYNU'].values ]
plt.figure(figsize=(25,8))
plt.bar(produkty_kategorie['IDPRODUKTU'], produkty_kategorie['STANMAGAZYNU'], color=colors)
plt.title('Stan magazynowy produktów w sklepie (czerwony kolor wskazuje poziom poniżej dopuszczalnego minimum)')
plt.xlabel('')
plt.ylabel('Ilość zamówień')
plt.xticks(produkty_kategorie['IDPRODUKTU'].values)
plt.xlim(0,len(produkty_kategorie['IDPRODUKTU'])+1)
# plt.xticks(rotation=90)

plt.grid()
plt.show()
```

ilość produktów po zamówieniu

```python
produkty_kategorie['MAGAZYN+ZAMOWIENIE'] = produkty_kategorie['STANMAGAZYNU']+produkty_kategorie['ILOSCZAMOWIONA']
```

```python
colors=[]
for i,r in produkty_kategorie.iterrows():
    if r['MAGAZYN+ZAMOWIENIE'] <=r['STANMINIMUM']:
        colors.append('r')
    else:
        colors.append('b')
plt.figure(figsize=(25,8))
plt.bar(produkty_kategorie['IDPRODUKTU'], produkty_kategorie['MAGAZYN+ZAMOWIENIE'], color=colors)
plt.title('Stan magazynowy produktów w sklepie po zamówieniu brakujących (czerwony kolor wskazuje poziom poniżej dopuszczalnego minimum)')
plt.xlabel('')
plt.ylabel('Ilość zamówień')
plt.xticks(produkty_kategorie['IDPRODUKTU'].values)
plt.xlim(0,len(produkty_kategorie['IDPRODUKTU'])+1)
# plt.xticks(rotation=90)

plt.grid()
plt.show()
```
